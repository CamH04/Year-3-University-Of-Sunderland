```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def main():
    print("-----------------")
    # Generate sine wave data
    x = np.linspace(0, 10, 100)
    y = np.sin(x)
    # Plot sine wave
    plt.plot()
    plt.xlabel()
    plt.ylabel()
    plt.title()
    plt.legend()
    plt.show()

if __name__ == "__main__":
    main()
```
